```python
#importing Descriptor and complaint type
df2 = pd.read_csv("/kaggle/input/ny-311-service-requests/311-service-requests-from-2010-to-present.csv", usecols=["Complaint Type", "Descriptor"])
df2.info(verbose=False, memory_usage="deep") 
```

```python

print(pd.isnull(df2).sum())
print(df2.info())
```

```python
#removing null values from discriptor

df2['Descriptor'] = df2['Descriptor'].fillna('Nothing')

#converting the column 1 to upper case
df2['Descriptor'] = df2['Descriptor'].apply(lambda x: x.upper())
```

```python
df2
```

```python
#converting the column 2 to upper case
df2['Complaint Type'] = df2['Complaint Type'].apply(lambda x: x.upper())
df2
```

```python
# Get the unique values in 'col1'
unique_values = df2['Complaint Type'].unique()

# Print the unique values
print(unique_values)
valuec=df2['Complaint Type'].value_counts()
print(len(valuec))

```

```python
#analysing the data of complaint type

complaint_type_counts = (
    df2.groupby(df2["Complaint Type"])
    .size()  # Count occurrences of each complaint type
    .to_frame(name="count")
    .reset_index()
    .sort_values(by="count", ascending=False)  # Sort by count descending
)

# Get top 10 complaint types
top_10_complaints = complaint_type_counts.head(10)

# Reorder complaint type labels based on count (descending)
complaint_order = top_10_complaints[df2["Complaint Type"]].tolist()
complaint_order = sorted(complaint_order, key=lambda x: top_10_complaints[top_10_complaints["Complaint Type"] == x]["count"].values[0], reverse=True)

# Create the bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed

plt.bar(complaint_order, top_10_complaints["count"])  # Use reordered labels

plt.xlabel("Complaint Type")
plt.ylabel("Count")
plt.title("Top 10 Complaint Types")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability

plt.tight_layout()
plt.show()
```

```python
#craeting a graph for analysis of complain type
plt.hist(df2["Complaint Type"],bins=149,edgecolor='black',rwidth=0.8)
#rotating the the labels of x axis by 90 degree
plt.xticks(rotation=90)
sns.despine(bottom=True) 
plt.xticks(fontsize=2) 
print(plt.figure(figsize=(8,6))  )#Figure(640x480)=original
plt.ylabel("FREQUENCY")
plt.xlabel("Complaint Type")
plt.legend()
plt.grid()
plt.show()
```

```python
#visualization 
# Assuming your data is in a pandas DataFrame named 'data'

# Borough analysis
borough_counts = (
    df4.groupby("Borough")
    .size()  # Count occurrences of each borough
    .to_frame(name="count")
    .reset_index()
    .sort_values(by="count", ascending=False)  # Sort by count descending
)

# Reorder borough labels based on count (descending)
borough_order = borough_counts["Borough"].tolist()
borough_order = sorted(borough_order, key=lambda x: borough_counts[borough_counts["Borough"] == x]["count"].values[0], reverse=True)

# Create the bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed

plt.bar(borough_order, borough_counts["count"])  # Use reordered labels for x-axis

plt.xlabel("Borough")
plt.ylabel("Count")
plt.title("Complaints by Borough")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability

plt.tight_layout()
plt.show()
```
